```python
#DB→ differential
def get_differential(SlidingTime, FrictionCoefficient):
  import pandas


  slope = []
  ave = [[]]
  delta = 200 #平均する微分の範囲
  for n in range(len(SlidingTime)-1):
    slope.append((FrictionCoefficient[n + 1] - FrictionCoefficient[n]) / (SlidingTime[n + 1] - SlidingTime[n]))
    if n%delta == delta - 1:
      for i in range(delta):
        ave_temp = (slope[n-(delta-1) + i] / delta)
      ave.append([SlidingTime[n-(delta-1)], ave_temp])
      ave_temp = 0
  return ave
```
