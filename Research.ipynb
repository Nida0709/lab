```python
#DB取得
def get_DB(data_Path):

  #Excel操作系ライブラリーのインポート
  import openpyxl
  wb_temp = openpyxl.load_workbook(data_Path)
  sheet_temp = wb_temp["Sheet1"]


  #Body
  #ファイルを開いてデータを取得
  outputDB = list(sheet_temp.rows)

  return outputDB
```

```python
#fileListから情報取得
def sprit_fileList(data_Path):

  import os

  #Body
  buf = os.path.basename(data_Path)

  countN = buf.find("_")
  Method = buf[ : countN]
  buf = buf[countN+1 : ]

  countN = buf.find("_")
  Material = buf[ : countN]
  buf = buf[countN+1 : ]

  countN = buf.find("_")
  Weight = buf[ : countN]
  buf = buf[countN+1 : ]

  countN = buf.find("_")
  SlidingSpeed = buf[ : countN]
  buf = buf[countN+1 : ]

  countN = buf.find(".")
  number = buf[ : countN]

  return Method, Material, Weight, SlidingSpeed, number
```
